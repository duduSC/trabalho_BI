```python
import pandas as pd
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv
movies_csv= "movies.csv"

movie_df= pd.read_csv(movies_csv,sep="|")

movie_df
```

```python

movie_df = movie_df.drop_duplicates(['id'])
movie_df = movie_df[~movie_df.isin([r"\N"])]
movie_df= movie_df.dropna()
movie_df
```

```python
movie_df = movie_df.drop(columns=["personagem","nomeArtista","anoNascimento","anoFalecimento","profissao","titulosMaisConhecidos","generoArtista"])
movie_df
```

```python
movie_df["genero"] = movie_df["genero"].str.split(",")
movie_df
```

```python
dim_filme = movie_df[["id","tituloPincipal","tituloOriginal"]]
dim_filme = dim_filme.rename(columns={"id":"id_imdb","tituloPincipal":"titulo_principal","tituloOriginal":"titulo_original"})
dim_filme
```

```python
fato_filme = movie_df[["id","anoLancamento","tempoMinutos","notaMedia","numeroVotos"]]
fato_filme = fato_filme.rename(columns={"id":"id_imdb","anoLancamento":"ano_lancamento","tempoMinutos":"tempo_minutos","notaMedia":"nota_media","numeroVotos":"numero_votos"})
fato_filme
```

```python
dim_genero= movie_df["genero"]
dim_genero = dim_genero.explode("genero")
dim_genero = dim_genero.rename("nome")
dim_genero = dim_genero.drop_duplicates()
dim_genero = pd.DataFrame(dim_genero)
dim_genero["id"]= range(len(dim_genero))
dim_genero
```

```python
df_explode = movie_df[["id","genero"]]
df_explode = df_explode.explode("genero")
df_explode= df_explode.rename(columns={"id":"id_imdb","genero":"nome"})

dim_filme["id_imdb"]= dim_filme.id_imdb.astype("str")
df_explode["id_imdb"] = df_explode.id_imdb.astype("str")
df_bridge = df_explode.merge(dim_genero,on="nome",how="inner")
df_bridge= df_bridge.rename(columns={"id":"id_genero"})

df_bridge
```

```python
load_dotenv()
DB_SENHA=os.getenv("DB_SENHA")
db = create_engine('postgresql://eduar:{DB_SENHA}@localhost:5432/DW_Filmes')
```

```python
dim_genero = dim_genero[["nome"]]
```

```python
dim_genero.to_sql("dim_genero",
                  con=db,
                  if_exists="delete_rows",
                  index=False)
dim_filme.to_sql("dim_filme",
                  con=db,
                  if_exists="delete_rows",
                  index=False)
```

```python
pd.read_sql("select * from dim_filme limit 10",db)
```

```python
pd.read_sql("SELECT * FROM dim_genero limit 10",db)
```

```python
df_bridge.to_sql("bridge_filme_genero",
                 con=db,
                 if_exists="replace",
                 index=False)

pd.read_sql("select * from bridge_filme_genero limit 10",db)
```

```python
query = "SELECT dim_genero.sk_genero, dim_filme.sk_filme FROM bridge_filme_genero INNER JOIN dim_filme ON bridge_filme_genero.id_imdb = dim_filme.id_imdb INNER JOIN dim_genero ON bridge_filme_genero.nome = dim_genero.nome"
df_bridge = pd.read_sql(query,db)
df_bridge.to_sql("bridge_filme_genero",
                 con=db,
                 if_exists="replace",
                 index=False)
pd.read_sql("select * from bridge_filme_genero limit 10",db)
```
